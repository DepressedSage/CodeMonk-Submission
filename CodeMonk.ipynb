```python
def make_data_vocab(df):
    df_dict = dict()
    for col in df.columns[1:]:
        unique_values = df[col].unique()
        df_dict[col] = {v: k for k, v in enumerate(unique_values)}
    return df_dict

def make_cat_vocab(df):
    df_dict = dict()
    for col in df.columns[1:]:
        unique_values = df[col].apply(lambda x: list(x.keys())[0]).unique()
        df_dict[col] = {v: k for k, v in enumerate(unique_values)}
    return df_dict

def tokenize_data(df,df_dict):
    tokenized_df = df.copy()

    for col in ['color', 'season', 'gender']:
        if col in df_dict:
            tokenized_df[col] = tokenized_df[col].apply(lambda x: df_dict[col][list(x.values())[0]] if isinstance(x, dict) else df_dict[col][x])
    return tokenized_df

def tokenize_cat(df,df_dict):
    tokenized_df = df.copy()

    for col in df_dict.keys():
        if col in tokenized_df.columns:
            tokenized_df[col] = tokenized_df[col].apply(lambda x: df_dict[col][list(x.keys())[0]])
    return tokenized_df
```
